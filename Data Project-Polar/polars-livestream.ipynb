```python
pokemon = pl.read_csv("C:/Users/lebac/Downloads/pokemon_data.csv")




```

```python
print(len(pokemon.columns))  # number of columns
```

```python
print(pokemon.filter(
    (pl.col("Legendary"))& (pl.col("Type 1")=="Grass")))
```

```python
print(pokemon.sample(5))
```

```python

#pl.selec(names,sorted  alphabetically ,Average Hp, max attack,speed)
#pokemon.columns 
print(pokemon.select(pl.col("Name").sort(),
    pl.col("HP").mean().alias("Average Hp"),#Of all pokeman
    pl.col("HP").max().alias("Max Attack"),#Of all pokeman
    pl.col("Speed")).sample(5))
```

```python
pokemon.filter(
    pl.col("HP")>159
)
```

```python
#Groupby Polars
```

```python
#Grab type 1 pokemon count
gen1 = pokemon.filter(
    pl.col("Generation")==1
)
out = gen1.group_by("Type 1").agg(
    pl.col("#").count().alias("Count")
)

out.select(
    pl.col("Type 1"),
    pl.col("Count")
).sort('Count', descending = True)
```
